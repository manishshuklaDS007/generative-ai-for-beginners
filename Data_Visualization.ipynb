```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# 1. Scatterplot: Relationship between age and fare
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('Scatterplot: Age vs Fare')
plt.show()

# 2. Lineplot: Trend of fare over passenger class
sns.lineplot(data=titanic, x='class', y='fare', hue='survived', ci=None)
plt.title('Lineplot: Fare Trends across Passenger Classes')
plt.show()

# 3. Barplot: Survival rate by class
sns.barplot(data=titanic, x='class', y='survived', ci=None)
plt.title('Barplot: Survival Rate by Class')
plt.show()

# 4. Histogram: Age distribution
sns.histplot(titanic['age'], bins=10, kde=True)
plt.title('Histogram: Age Distribution')
plt.show()

# 5. Boxplot: Fare distribution by class
sns.boxplot(data=titanic, x='class', y='fare')
plt.title('Boxplot: Fare by Class')
plt.show()

# 6. Violin Plot: Age distribution by sex
sns.violinplot(data=titanic, x='sex', y='age', hue='survived')
```

# a. Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg, hp-time-to-60) with respect to each brand in the same plane

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/cars (1).csv")
```

```python
df
```

```python
# a. Create scatter plots with between  with respect to each brand in the same plane
(mpg-hp, mpg-weight), and (hp-mpg, hp-time-to-60)
```

```python
plt.figure(figsize=(14,10))
```

```python
#mpg vs hp
plt.subplot(2,2,1)
sns.scatterplot(data=df,x='hp',y='mpg',hue='brand')
plt.title("mpg vs hp")

#mpg-weight
plt.subplot(2,2,2)
sns.scatterplot(data=df,x='weightlbs',y='mpg',hue='brand')
plt.title("mpg-weight")

#hp-mpg
plt.subplot(2,2,3)
sns.scatterplot(data=df,x='mpg',y='hp',hue='brand')
plt.title("hp vs mpg")

#hp-time-to-60
plt.subplot(2,2,4)
sns.scatterplot(data=df,x='hp',y='time-to-60',hue='brand')
plt.title("hp  vs time-to-60")

plt.tight_layout()
plt.show()
```

Create box plots for mpg, hp, weight, time-to-60 vs cylinders and with respect to brand

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.boxplot(x='cylinders', y='mpg', hue='brand', data=df)
plt.title('MPG vs. Cylinders by Brand')
plt.subplot(2, 2, 2)
sns.boxplot(x='cylinders', y='hp', hue='brand', data=df)
plt.title('HP vs. Cylinders by Brand')
plt.subplot(2, 2, 3)
sns.boxplot(x='cylinders', y='weightlbs', hue='brand', data=df)
plt.title('Weight vs. Cylinders by Brand')
plt.subplot(2, 2, 4)
sns.boxplot(x='cylinders', y='time-to-60', hue='brand', data=df)
plt.title('Time-to-60 vs. Cylinders by Brand')
plt.tight_layout()
plt.show()
```
